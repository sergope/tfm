```
## Extracción de un csv para una Estación en concreto. Filtrado por municipio

def validate_inputs(years, months, station_name):
  if isinstance(years, list) and isinstance(months, list) and isinstance(station_name, str):
    return True
  else:
    print('years and months must be a list[] and station must be a string')
    print('years is a list: ' + str(isinstance(years, list)))
    print('months is a list: ' + str(isinstance(months, list)))
    print('station is a string: ' + str(isinstance(station_name, str)))
    return False



def get_station_by_name(station_name):
  %run 'drive/My Drive/TFM/resources/enums'.ipynb
  try:
    station = estacion_de_medida[station_name.upper()]
    return station
  except:
    print('station: "' + station_name + '" does not exists')



def extract_data_by_station(years, months, station_name):
  if not validate_inputs(years, months, station_name):
    return

  station_data = get_station_by_name(station_name)
  if not station_data:
    return
  
  all_years = []
  for year in years:
    filtered_months = []
    for month in months:
      file_name = str(year) + month + '.csv'
      try:
        table = pd.read_table(BASE_FOLDER+'data/' + str(year) + '/' + file_name, sep=';', engine='python')
        table.columns = table.columns.str.upper()
        filtered_months.append(table[table['MUNICIPIO'] == station_data['municipio']])
      except:
        print('File: "' + BASE_FOLDER+'data/' + str(year) + '/' + file_name + '" does not exists')
        pass

    all_years.append(pd.concat(filtered_months))
    print('Year ' + str(year) + ' loaded')

  final_table = pd.concat(all_years)
  final_table.to_csv(BASE_FOLDER + 'processed_data/' +  station_name.lower() + '.csv', sep = ';', na_rep = 'null')
  print('Process finished. File path: ' + BASE_FOLDER + 'processed_data/' +  station_name.lower() + '.csv')
```

```
#Generamos el csv de GETAFE

years = np.arange(2005,2020).tolist()  #desde el año 2005 hasta 2020
months = ['_01', '_02', '_03', '_04', '_05', '_06', '_07', '_08', '_09', '_10', '_11', '_12']
station = 'getafe'

extract_data_by_station(years, months, station)
```
